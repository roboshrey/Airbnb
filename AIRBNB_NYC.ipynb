# Data Cleaning


```python
# droping the name and last review coulumn as they are not gonna give usefull insights


df.drop(['name'], axis = 1, inplace= True)
```

```python
df.drop(['last_review'], axis = 1, inplace= True)
```

```python
# replacing the null values with o and etc for cleaning the dataset

df['reviews_per_month'].replace(np.nan, 0 , inplace= True)
df['host_name'].replace(np.nan, 'Not_Provided' , inplace= True)
```

```python
df.info()
```

```python
df.describe()
```

```python
# minimum price is showing 0 which is corrupt data, hence mending the dataset with the rows that doest have 0 as minimum price

df=df.loc[df['price']>0]
df.describe()
```

```python
# Adding the dollar sign infront of price column to make clear understanding

def add_dollars(value):
    value = str(value) + "$"
    return value

df['price'] = df['price'].apply(add_dollars)

df['price']
```

# Data Visualization


```python
# Here lets see the top host in the dataset

top_host=df['host_name'].value_counts()[:5]
top_host
```

```python


plt.figure(figsize=(9, 3))
sns.barplot(x=df['host_name'].value_counts()[:5].index,
            y=df['host_name'].value_counts()[:5].values,
            palette='viridis')
plt.title("Top host name", fontsize=15)
plt.xlabel('Host', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

So we categorized the group by the count of host name and took up the top 5 names , we found that Michel followed by David and others were the highest host in order of booking rooms

```python
# Which room type is mostly picked up
room_type_count=df['room_type'].value_counts().plot(kind='pie',figsize=(5,5),autopct='%1.1f%%', fontsize=15)
plt.title('Room Type count', fontsize=20)
plt.show()
```

```python
# expensive neighbourhood group

plt.figure(figsize=(10, 7))
sns.countplot(x='neighbourhood_group', hue='room_type', data=df, palette='bright')
plt.title('Room Type in Neighbourhood Groups', fontsize=15)
plt.show()
```

```python
reviews_area= df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
reviews_area=reviews_area.sort_values(by='number_of_reviews', ascending=False)
reviews_area
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x='neighbourhood_group',
            y='number_of_reviews',
            data= reviews_area)
plt.title("Reviews by area", fontsize=15)
plt.xlabel('Group', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

```python
highest_prize= df.groupby(['host_id','neighbourhood_group','host_name'])['price'].max().reset_index()
highest_prize=highest_prize.sort_values(by='price',ascending=False).head(5)
highest_prize

```

```python

plt.figure(figsize=(9, 5))
sns.barplot(x='neighbourhood_group',
            y='price',
            data=highest_prize,
            palette='husl')
plt.title("Top 5 highest prices", fontsize=15)
plt.xlabel("Neighbourhood Group", fontsize=9)
plt.ylabel("Price", fontsize=9)
plt.show()
```
